```python
map = folium.Map(location=[31.2, 76.5], zoom_start=12)

g1 = folium.FeatureGroup(name='Kiratpur Sahib')
map.add_child(g1)

g2 = folium.FeatureGroup(name='Chamkaur Sahib')
map.add_child(g2)

g3 = folium.FeatureGroup(name='City Ropar')
map.add_child(g3)

g4 = folium.FeatureGroup(name='Morinda')
map.add_child(g4)

g5 = folium.FeatureGroup(name='Nurpur Bedi')
map.add_child(g5)

g6 = folium.FeatureGroup(name='Sadar Roopnnagar')
map.add_child(g6)

g7 = folium.FeatureGroup(name='Singh Bhagwantpura')
map.add_child(g7)

g8 = folium.FeatureGroup(name='Anandpur Sahib')
map.add_child(g8)

g9 = folium.FeatureGroup(name='Nangal')
map.add_child(g9)


#for point in range(0, len(df)):
  #  if (point<26):
 #       folium.Marker(locationlist[point], popup=KS['Police Station'][point],tooltip=KS['Place/Location of Occurrence'][point]).add_to()



for point in range(0, len(df)):
    if (point<26):
        folium.Marker(locationlist[point], popup=KS['Police Station'][point],tooltip=KS['Place/Location of Occurrence'][point], icon=folium.Icon(color='lightgray')).add_to(g1)

        
    if(point>26 and point<40):
        folium.Marker(locationlist[point], popup=CS['Police Station'][point],tooltip=CS['Place/Location of Occurrence'][point],icon=folium.Icon(color='black')).add_to(g2)

    
    if (point>=41 and point<76):
        folium.Marker(locationlist[point], popup=CR['Police Station'][point],tooltip=CR['Place/Location of Occurrence'][point],icon=folium.Icon(color='darkblue')).add_to(g3)
        
        
    if (point>=76 and point<102):
        folium.Marker(locationlist[point], popup=MR['Police Station'][point],tooltip=MR['Place/Location of Occurrence'][point],icon=folium.Icon(color='orange')).add_to(g4)
        
        
    if (point>=102 and point<116):
        folium.Marker(locationlist[point], popup=NB['Police Station'][point],tooltip=NB['Place/Location of Occurrence'][point],icon=folium.Icon(color='blue')).add_to(g5)
        
        
    if (point>=116 and point<141):
        folium.Marker(locationlist[point], popup=SR['Police Station'][point],tooltip=SR['Place/Location of Occurrence'][point],icon=folium.Icon(color='beige')).add_to(g6)
        
        
    if (point>=141 and point<162):
        folium.Marker(locationlist[point], popup=SB['Police Station'][point],tooltip=SB['Place/Location of Occurrence'][point],icon=folium.Icon(color='purple')).add_to(g7)
        
        
    if (point>=162 and point<188):
        folium.Marker(locationlist[point], popup=AS['Police Station'][point],tooltip=AS['Place/Location of Occurrence'][point],icon=folium.Icon(color='green')).add_to(g8)
        
        
    if (point>=188 and point<217):
        folium.Marker(locationlist[point], popup=NG['Police Station'][point],tooltip=NG['Place/Location of Occurrence'][point],icon=folium.Icon(color='cadetblue')).add_to(g9)
    
folium.LayerControl(collapsed=False).add_to(map)
map.save('map.html')
map
```
