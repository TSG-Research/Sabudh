```python
import folium                    
import folium.plugins as plugins
from folium import plugins
from folium.plugins import HeatMapWithTime
import pandas as pd
```

```python
ropar_analysis='ropar_analysis.xlsx'
dfs=pd.read_excel(ropar_analysis)
dfs.head(5)
```

```python
Ruraldf=dfs[dfs['Rural/Urban']=='Rural']

Ruraldf=pd.DataFrame(Ruraldf, columns=['Police Station','Place/Location of Occurrence','Latitude','Longitude','Rural/Urban'])


Urbandf=dfs[dfs['Rural/Urban']=='Urban']

Urbandf=pd.DataFrame(Urbandf, columns=['Police Station','Place/Location of Occurrence','Latitude','Longitude','Rural/Urban'])

#Ruraldf.reset_index(inplace = True, drop = True) 
Urbandf.head(5)
```

```python
#grp = dfs.groupby('Rural/Urban')

#U=grp.get_group('Urban')
#R=grp.get_group('Rural')
#len(R)

Ruraldf.reset_index(inplace = True, drop = True) 

Ruraldf.head(5)

Urbandf.reset_index(inplace = True, drop = True) 

Urbandf.head(5)
```

```python
Urbandf.head(5)
```

```python
locationsR = Ruraldf[['Latitude', 'Longitude']]
locationlistR = locationsR.values.tolist()
len(locationlistR)
locationlistR[7]

locationsU = Urbandf[['Latitude', 'Longitude']]
locationlistU = locationsU.values.tolist()
len(locationlistU)

```

```python
map = folium.Map(location=[31.2, 76.5], zoom_start=12)


g1 = folium.FeatureGroup(name='Rural')
map.add_child(g1)

g2 = folium.FeatureGroup(name='Urban')
map.add_child(g2)


for point in range(0, len(Ruraldf)):
    folium.Marker(locationlistR[point], popup=Ruraldf['Police Station'][point],tooltip=Ruraldf['Place/Location of Occurrence'][point], icon=folium.Icon(color='blue')).add_to(g1)

for point in range(0, len(Urbandf)):
    folium.Marker(locationlistU[point], popup=Urbandf['Police Station'][point],tooltip=Urbandf['Place/Location of Occurrence'][point], icon=folium.Icon(color='red')).add_to(g2)



folium.LayerControl(collapsed=False).add_to(map)
map.save('map_Urban_Rural.html')
map     
```
